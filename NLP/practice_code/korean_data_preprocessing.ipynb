## Text Preprocessing

```
punct = "/-'?!.,#$%\'()*+-/:;<=>@[\\]^_`{|}~" + '""“”’' + '∞θ÷α•à−β∅³π‘₹´°£€\×™√²—–&'
punct_mapping = {"‘": "'", "₹": "e", "´": "'", "°": "", "€": "e", "™": "tm", "√": " sqrt ", "×": "x", "²": "2", "—": "-", "–": "-", "’": "'", "_": "-", "`": "'", '“': '"', '”': '"', '“': '"', "£": "e", '∞': 'infinity', 'θ': 'theta', '÷': '/', 'α': 'alpha', '•': '.', 'à': 'a', '−': '-', 'β': 'beta', '∅': '', '³': '3', 'π': 'pi', }
```

```
# remove special words in each sentences

def clean_punc(text, punct, mapping):
  
  for i in mapping:
    text = text.replace(i, mapping[i])

  for i in punct:
    text = text.replace(i, f' {i} ')
  
  specials = {'\u200b': ' ', '…': ' ... ', '\ufeff': '', 'करना': '', 'है': ''}
  for i in specials:
    text.replace(i, specials[i])

  return text.strip()
```

```
cleaned_corpus = []
for text in korean_text:
  cleaned_corpus.append(clean_punc(text, punct, punct_mapping))
```

```
import re

def remove_specials(texts):
  corpus = []
  print(len(texts))
  for i in range(0, len(texts)):
    review = re.sub(r'[@%\\*=()/~#&\+á?\xc3\xa1\-\|\.\:\;\!\-\,\_\~\$\'\"]', '',str(texts[i]))
    review = re.sub(r'\d+','', review)# remove number
    review = review.lower() #lower case
    review = re.sub(r'\s+', ' ', review)
    review = re.sub(r'<[^>]+>','',review) #remove Html tags
    review = re.sub(r'\s+', ' ', review) #remove spaces
    review = re.sub(r"^\s+", '', review) #remove space from start
    review = re.sub(r'\s+$', '', review) #remove space from the end
    corpus.append(review)

  return corpus
```

```
basic_preprocessed_corpus = remove_specials(cleaned_corpus)
```

**Import libraries**

```
from pykospacing import spacing
from hanspell import spell_checker
from soynlp.normalizer import *
```

```
# replace wrong foreign words

lownword_map = {}
lownword_data = open('/content/confused_loanwords.txt','r',encoding='utf-8')

lines = lownword_data.readlines()

for line in lines:
  line = line.strip()
  miss_spell = line.split('\t')[0]
  ori_word = line.split('\t')[1]
  lownword_map[miss_spell] = ori_word
```

## Modify a grammatically incorrect sentence

```
def spell_check_text(texts):
  corpus = []
  for text in texts:
    spacing_text = spacing(text)
    spelled_text = spell_checker.check(spacing_text)
    checked_text = spelled_text.checked
    normalized_text = repeat_normalize(checked_text)
    for lownword in lownword_map:
      normalized_text = normalized_text.replace(lownword, lownword_map[lownword])
    corpus.append(normalized_text)
  return corpus
```

```
spell_preprocessed_corpus = spell_check_text(basic_preprocessed_corpus)
```

```
from khaiii import KhaiiiApi
api = KhaiiiApi()
```

```
word = "나도 모르게 사버렸다."
print(type(word))
for w in api.analyze(word):
  for morph in w.morphs:
    print(morph.lex)
print(spell_preprocessed_corpus[0])
```

```
def pos_text(texts):
  corpus = []
  for text in texts:
    if text == '':
      continue
    for word in api.analyze(text):
      for morph in word.morphs:
        corpus.append(morph.lex)
        # print(morph.lex)
  return corpus    
```

```
pos_khaiii_corpus = pos_text(spell_preprocessed_corpus)
```

**make txt file for SentencePiece**

```
f = open('spm_practce.txt','w')
for text in spell_preprocessed_corpus:
  if text == '':
    continue
  else:
    f.write(text+'\n')
f.close() 
```

```
templates = '--input={} \
--pad_id={} \
--bos_id={} \
--eos_id={} \
--unk_id={} \
--model_prefix={} \
--vocab_size={} \
--character_coverage={} \
--model_type={}'

train_input_file = '/content/spm_practce.txt'
pad_id = 0
bos_id = 1
eos_id = 2
unk_id = 3
prefix = 'spm_model'
vocab_size = 2023 # change by your dataset
character_coverage = 1
model_type = 'word'

cmd = templates.format(train_input_file,pad_id,bos_id,eos_id,unk_id,prefix,vocab_size,character_coverage,model_type)
```

**make sentencepiece model, vacab**

```
import sentencepiece as spm

spm.SentencePieceTrainer.Train(cmd)
sp = spm.SentencePieceProcessor()
sp.load('spm_model.model')
```

```
print(sp.get_piece_size())
print(sp.id_to_piece(503))
print(sp.piece_to_id('▁가루에'))
```

```
# add <s>, </s> token in sentences
sp.SetEncodeExtraOptions('bos:eos')
```

**문자열 → 토큰화**

```
# 100문장으로 vocab을 만들어서 중간 중간 <unk>가 있습니다.
tokens = sp.EncodeAsPieces('나는 그대를 뒤로 한 채로 우리 사랑을 놓아요.')
print(tokens)
tokens2 = sp.EncodeAsIds('나는 그대를 뒤로 한 채로 우리 사랑을 놓아요.')
print(tokens2)
```

**토큰 → 문자화**

```
token = sp.DecodePieces(tokens)
print(token)
token2 = sp.DecodeIds(tokens2)
print(token2)
```
